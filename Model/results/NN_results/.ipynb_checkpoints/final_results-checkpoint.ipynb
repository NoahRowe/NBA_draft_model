```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
money_per_win = 3.03 # Million dollars

counts = list(range(1, 31))
data = []
for c in counts:
        path = "standard_3_manhattan_{}.npy".format(c)
        data.append(np.load(path, allow_pickle=True).tolist())

# organize it
mses, new_picks, old_picks, new_wins, team_wins = [], [], [], [], [[] for i in range(30)]
actual, preds = [], []
for d in data:
    mses.append(d["MSE"])
    new_picks.append(np.mean(d["new_picks"]))
    old_picks.append(np.mean(d["old_picks"]))
    new_wins.append(np.mean(d["new_added_wins"]))
    actual += d['actual_values'].tolist()
    preds += d['predictions'].tolist()
    for i in range(30):
        team_wins[i].append(d['added_wins_per_team'][i])
team_avg, team_std = [np.mean(val) for val in team_wins], [np.std(val) for val in team_wins]
```

```python
# Average mean squared error
print("MSE: {:.3f} +/- {:.4f}".format(np.mean(mses), np.std(mses)))
# Average wins of old picks and new picks
print("New Picks: {:.3f} +/- {:.4f}".format(np.mean(new_picks), np.std(new_picks)))
print("Old Picks: {:.3f} +/- {:.4f}".format(np.mean(old_picks), np.std(old_picks)))
# Average difference
print("Average Difference: {:.3f} +/- {:.4f}".format(np.mean(new_wins), np.std(new_wins)))
# Average team benefit
print("Average team benefit: {:.3f} +/- {:.3f}".format(np.mean(team_avg), np.std(team_avg)))
```

```python
# Chances that a pick is worse off
print("Percent of Worse Picks: {:.4f}%".format(sum(np.array(new_wins)<0)/len(new_wins)*100))
sum(np.array(new_wins)<0)
```

```python
# Plot the average new wins of each team
labels = ['ATL', 'BOS', 'BRK', 'CHI', 'CHO', 'CLE', 'DAL', 'DEN', 'DET', 'GSW', 'HOU', 'IND',
 'LAC', 'LAL', 'MEM', 'MIA', 'MIL', 'MIN', 'NOP', 'NYK', 'OKC', 'ORL', 'PHI', 'PHO',
 'POR', 'SAC', 'SAS', 'TOR', 'UTA', 'WAS']
plt.figure(figsize=(10,8), dpi=100)
avgHeight, stdHeight = np.mean(team_avg), np.std(team_avg)
x = range(30)
plt.bar(x, team_avg, edgecolor="black")
plt.plot([min(x), max(x)], [avgHeight, avgHeight], "--r", linewidth=3, 
         label="Average ({:.2f} +/- {:.2f})".format(avgHeight, stdHeight))
plt.legend()
plt.ylabel("Added Wins Made")
#plt.yticks(np.arange(-1.5, 3.5, 0.5))
plt.xticks(x, labels, rotation=85)
plt.grid(axis="y")
plt.show()

# DO IT FOR MONEY NOW
plt.figure(figsize=(10,8), dpi=100)
team_money_avg = [val*money_per_win for val in team_avg]
print(np.mean(team_avg), np.std(team_avg))
print(np.mean(team_money_avg), np.std(team_money_avg))
x = range(30)
plt.bar(x, team_money_avg, edgecolor="black")
plt.plot([min(x), max(x)], [avgHeight, avgHeight], "--r", linewidth=3, 
         label="Average ({:.2f} +/- {:.2f})".format(avgHeight, stdHeight))
plt.legend()
plt.ylabel("Added Wins Per Pick in Millions of Dollars")
#plt.yticks(np.arange(-1.5, 3.5, 0.5))
plt.xticks(x, labels, rotation=85)
plt.grid(axis="y")
plt.show()
```

```python
import matplotlib
matplotlib.rcParams.update({'font.size': 16})
fig, ax1 = plt.subplots(figsize=(12,8))

ax1.set_ylabel('Added Wins Per Draft Pick')

ax1.bar(x, team_avg, edgecolor="black")
ax1.plot([min(x), max(x)], [avgHeight, avgHeight], "--r", linewidth=3, 
         label="Average ({:.2f} +/- {:.2f})".format(avgHeight, stdHeight))

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

ax2.set_ylabel('Millions of Dollars')  # we already handled the x-label with ax1
ax2.bar(x, team_money_avg, edgecolor="black", alpha=0)

ax2.set_xticks(x)
ax2.set_xticklabels(labels)
ax1.tick_params(axis='x', labelrotation=45)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```python
plt.figure(figsize=(6.5,4))
width = 0.15       # the width of the bars
r1 = width/2.
r2 = -width/2.

plt.bar(r2, [np.mean(old_picks)], width=width, label="Old Wins per Draft Pick", edgecolor="black")
plt.bar(r1, [np.mean(new_picks)], width=width, label="New Wins per Draft Pick", edgecolor="black")

plt.ylabel("Wins per Draft Pick")
plt.xticks([-width/2, width/2], ['Historic Selections', 'Model Selections'])

# plt.legend(
#     loc="lower right"
# )
plt.show()
```

```python
# Compare actual and prediction distributions
scaledDraft = pd.DataFrame({"WM":actual, "WM_pred":preds})
plt.figure(figsize=(7,5), dpi=100)
plt.title("Violin Plot of Wins Made")
plt.grid()
plotData = scaledDraft[["WM", "WM_pred"]].melt().assign(x="")
sns.violinplot(data=plotData, x="x", y="value", hue="variable", inner="quartile", split=True)
plt.ylabel("Wins Made")
plt.xlabel("")
plt.show()
```
